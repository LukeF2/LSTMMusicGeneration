```python
from mido import MidiFile
import os

def count_midi_data_points(midi_folder):
    results = {}
    total_summary = {'Note On Events': 0, 'Note Off Events': 0, 'Control Change Events': 0, 'Total Events': 0}
    for file in os.listdir(midi_folder):
        if file.endswith('.mid'):
            midi_path = os.path.join(midi_folder, file)
            try:
                midi = MidiFile(midi_path)
                note_on_events = 0
                note_off_events = 0
                control_change_events = 0
                for track in midi.tracks:
                    for msg in track:
                        if msg.type == 'note_on':
                            note_on_events += 1
                        elif msg.type == 'note_off':
                            note_off_events += 1
                        elif msg.type == 'control_change':
                            control_change_events += 1
                total_events = note_on_events + note_off_events + control_change_events
                results[file] = {
                    'Note On Events': note_on_events,
                    'Note Off Events': note_off_events,
                    'Control Change Events': control_change_events,
                    'Total Events': total_events
                }
                total_summary['Note On Events'] += note_on_events
                total_summary['Note Off Events'] += note_off_events
                total_summary['Control Change Events'] += control_change_events
                total_summary['Total Events'] += total_events
            except Exception as e:
                print(f"Error processing {file}: {e}")
    results['Total Summary'] = total_summary
    return results

# Your MIDI folder path
midi_folder = "/Users/lukefeng/Documents/GitHub/Big Data Project/Chopin Midis"

# Counting the data points for all MIDI files in the folder
data_points = count_midi_data_points(midi_folder)

for file_name, counts in data_points.items():
    print(f"{file_name}:")
    for event_type, count in counts.items():
        print(f"  {event_type}: {count}")
    print()  # Blank line for readability



```

```python
from music21 import converter, note, chord
import os

def analyze_midi_files(folder_path):
    total_notes_and_chords = 0  # Initialize a counter for the total number of notes and chords

    for file_name in os.listdir(folder_path):
        if file_name.endswith('.mid'):
            midi_path = os.path.join(folder_path, file_name)
            try:
                # Load the MIDI file
                midi = converter.parse(midi_path)
                notes_to_parse = midi.flat.notes
                
                print(f"Analyzing {file_name}:")
                print(f"Total notes and chords: {len(notes_to_parse)}")
                total_notes_and_chords += len(notes_to_parse)  # Add to the total count
                
                # Optionally, print details for the first few notes/chords
                for element in notes_to_parse[:5]:
                    if isinstance(element, note.Note):
                        print(f"Note: {element.pitch}, Octave: {element.octave}, Offset: {element.offset}")
                    elif isinstance(element, chord.Chord):
                        print(f"Chord: {'+'.join(str(n) for n in element.normalOrder)}, Offset: {element.offset}")
                print()  # Add a blank line for readability
                
            except Exception as e:
                print(f"Error processing {file_name}: {e}")

    # After all files have been processed, print the total count
    print(f"Total notes and chords across all files: {total_notes_and_chords}")

# Specify the folder containing your MIDI files
midi_folder = "/Users/lukefeng/Documents/GitHub/Big Data Project/Romantic Midis copy"
 # Update this path to your specific folder

# Analyze the MIDI files in the specified folder
analyze_midi_files(midi_folder)
```
